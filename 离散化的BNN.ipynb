```python
df_1 = read_excel("4-fenwei.xlsx")
df_1.head()
```

```python
df_1.hist(bins=10,figsize=(16,14))
```

```python
from sklearn.model_selection import train_test_split
# 训练集的特征值x_train 测试集的特征值x_test 训练集的目标值y_train 测试集的目标值y_test，指定测试集所占的比例为20%
x_train, x_test, y_train, y_test = train_test_split(df_1, df.label, random_state=22,test_size=0.2)
print("x_train:\n", x_train.shape)
print("x_test:\n",x_test.shape)
x_train = x_train.reset_index(drop=True)  #重设索引
x_train
```

```python
from pgmpy.models import BayesianModel
from pgmpy.estimators import BayesianEstimator
import warnings

with warnings.catch_warnings():
    warnings.simplefilter('ignore')

    model = BayesianModel([('label', 'X1'), ('label', 'X2'),('label','X3'),('label','X4'),('label','X5'),('label', 'X6'), ('label', 'X7'),('label','X10'),('label','X11'),('label','X12'),('label', 'X13'), ('label', 'X14'),('label','X15'),('label','X16'),('label','X17'),('label','X18'),('label','X19')])
```

```python
from graphviz import Digraph

save=False
node_attr = dict(
    style='filled',
    shape='box',
    align='left',
    fontsize='10',
    ranksep='0.1',
    height='0.2'
)
dot = Digraph(node_attr=node_attr, graph_attr=dict(size="10,10"))
seen = set()
edges=model.edges()
for a,b in edges:
    dot.edge(a,b)
if save:
    dot.view(cleanup=True)
dot
```

```python
model.fit(data = x_train, estimator=BayesianEstimator)
model
```

```python
#输出节点信息
#print(model.nodes())
#输出依赖关系
#print(model.edges())
#查看某节点概率分布
print(model.get_cpds('X7').values)
```

```python
import pandas as pd

y_test = pd.DataFrame(y_test,columns=['label'])
y_test1 = y_test.reset_index(drop=True)  #重设索引
#x_test1 = x_test.reset_index(drop=True)  #重设索引
predict_data=x_test.drop(columns=['label'],axis=1)
y_pred = model.predict(predict_data)
print((y_pred['label']==y_test1['label']).sum()/len(x_test))
```
